```python
import numpy as np

# Load the image shape and array from the text file
with open('image_array.txt', 'r') as f:
    height, width = map(int, f.readline().split())
    img_array = np.loadtxt(f, dtype=int)

if img_array.shape[1] == 3:
    # RGB image
    img = img_array.reshape(height, width, 3)
else:
    # Grey Scale image
    img = img_array.reshape(height, width)

# Display the loaded image
import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()
```
